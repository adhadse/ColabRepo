# 8.2 Combining and Merging Datasets

### Many-to-One merge

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
# data2 has multiple rows same data where the key were overlapping.
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, how='outer')
```

### Many-to-Many merge

Cartesian product

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
df1
```

```python
df2
```

```python
# Many-to-may joins form the cartesian product
# 3 rows of key'b' in df1 and 2 in df2; resulting in 6 rows in merge
pd.merge(df1, df2,  on='key', how='left')
```

```python
pd.merge(df1, df2, how='inner')
```

```python
# To merge with multiple keys
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4 , 5, 6, 7]})
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
left
```

```python
right
```

```python
# Overlapping columns with same name but not included in merge operation
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=['_left', '_right'])
```

# 8.3 Reshaping and Pivoting

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                    index=pd.Index(['Ohio', 'Colorado'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],name='number'))
data
```

```python
# using the stack method on this data pivots the columns into rows and produce a Series
result = data.stack()
result
```

```python
#By default innermost level is stacked/Unstacked.
result.unstack()
```

```python
# By passing a level number or name, unstack different level
# But cannot stack different level other than innermost level
result.unstack(0)
```

```python
result.unstack('state')
```

```python
result = data.stack('number')
result
```

```python
# Unstacking might introduce missing data if all the values in the level  aren't found in each of the subgroups

s1 = pd.Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([4, 5, 6], index=['c', 'd', 'e'])
data2 = pd.concat([s1, s2,], keys=['one', 'two'])
data2
```

```python
data2.unstack()
```

```python
data2.unstack().stack()
```

```python
data2.unstack().stack(dropna=False)
```

### Unstack in a dataframe, the level unstacked becomes the lowest level in the result

```python
df = pd.DataFrame({'left': result, 'right': result + 5},
                  columns=pd.Index(['left', 'right'], name='side'))
df                  
```

```python
df.unstack('state')
```

```python
# When calling stack, we can indicate the name of the axis to stack
df.unstack('state').stack('side')
```

## Pivoting "Long" to " Wide" format

```python
data = pd.read_csv('datasets/macrodata.csv')
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year, 
                         quarter=data.quarter,
                         name='date')
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D', 'end')
ldata = data.stack().reset_index().rename(columns={0: 'value'})
```

### Long Format
Data is usually stored like this in "long-format" in db. But it might be difficult to work in this form.



```python
ldata[:10]
```

You may actually want a column for each item vaue indexed by date. **Pivot it**.
First two vales are the column to use as ***row index***, ***column index***

Finally optional ***value column*** to fill dataframe

```python
pivoted = ldata.pivot('date', 'item', 'value')
pivoted
```

### Suppose you wanted to reshape simulatenously

```python
ldata['value2'] = np.random.randn(len(ldata))
ldata[:3]
```

```python
# By omitting last argument, you get df with hierarchical columns
pivoted = ldata.pivot('date', 'item')
pivoted[:3]
```

```python
unstacked = ldata.set_index(['date', 'item']).unstack('item')
unstacked[:3]
```

## Pivoting "Wide" to "Long" Format

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3], 
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
```

### When using pd.melt, we must indicate which columns are group indicator

```python
melted = pd.melt(df, ['key'])
melted
```

```python
# using pivot, we can reshpae back to original lasyout
reshaped = melted.pivot('key', 'variable', 'value')
reshaped.columns.name = None
reshaped
```

```python
# use reset_index() to move data back into a column
reshaped.reset_index()
```

### You can also specify a subset of column to use as value columns

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
```

### Can be also used without any group identifiers

```python
pd.melt(df, value_vars=['A', 'B', 'C'])
```

```python
pd.melt(df, value_vars=['key', 'A', 'B', 'C'])
```
