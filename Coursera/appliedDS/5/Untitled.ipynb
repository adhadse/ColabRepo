```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib notebook
G = nx.Graph()
G.add_edge('a','b')
G.add_edge('b','d')
G.add_edge('a','c')
G.add_edge('c','d')
G.add_edge('c','e')
G.add_edge('e','d')
G.add_edge('e','g')
G.add_edge('g','d')
G.add_edge('g','f')
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx(G,pos)
```

```python
nx.degree_centrality(G)['d']
```

```python
nx.closeness_centrality(G)['g']
```

```python
nx.betweenness_centrality(G,endpoints= False)
```

```python
nx.edge_betweenness_centrality(G,normalized=False)
```

```python
B = nx.Graph()
B.add_edge('c','g')
B.add_edge('c','a')
B.add_edge('a','d')
B.add_edge('g','d')
B.add_edge('b','d')
B.add_edge('a','e')
B.add_edge('d','e')
B.add_edge('d','h')
B.add_edge('e','h')
B.add_edge('h','f')
pos1 = nx.spring_layout(B)
nx.draw_networkx(B,pos1)
plt.tight_layout()
```

```python
print(len(list(nx.common_neighbors(B,'h','e'))))
print(len(list(nx.common_neighbors(B,'h','f'))))
print(len(list(nx.common_neighbors(B,'h','d'))))
print(len(list(nx.common_neighbors(B,'h','b'))))
print(len(list(nx.common_neighbors(B,'h','g'))))
print(len(list(nx.common_neighbors(B,'h','a'))))
print(len(list(nx.common_neighbors(B,'h','c'))))



```

```python
common_neighbors = [(e[0],e[1],len(list(nx.common_neighbors(B,e[0],[1])))) for e in nx.non_edges(B)]
print(sorted(nx.common_neighbors))
```
